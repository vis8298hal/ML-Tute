# Case Study : Telecom Data Analyze Customer Calling Behaviour

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Datasets/telecom_customer_data.csv")
df.head()
```

## Relational Plot : Call Duration Vs Satisfaction Score

```python
#Create Figure 
plt.figure(figsize=(8,5))
#Plot Scatterplot
sns.scatterplot(data=df, x="Call_Duration", y="Satisfaction_Score", hue="Network_Provider")
plt.legend(title="Call Duration Vs Satisfaction Score", bbox_to_anchor=(1,1), loc="upper left")
#Label the Graph
plt.xlabel("Call Duration")
plt.ylabel("Satisfaction Score")
plt.title("Call Duration Vs Satisfaction Score")
#Show the Plot
plt.show()
```

##### Inference
- When the Call duration is low the satisfaction Score is high
- When the Call duration is high the satisfaction score is low.

## Categorical Plot: Count of Customers by Network Provider

```python
plt.figure(figsize=(6,4))
sns.countplot(data = df, x="Network_Provider", hue = "Network_Provider")
plt.xlabel("Network Providers")
plt.ylabel("Customer Count")
plt.title("Count of Customers by Network Provider")

plt.show()
```

#### Inference
- Provider A is having Highest number of customers.
- Provider C is having Least number of Customers.
- Difference between Customer Numbers between two proividers is 10%

## Distribution Plot : Distribution of Call Duration Per Customer 

```python
plt.figure(figsize=(7,5))
sns.histplot(df['Call_Duration'], bins=20, kde=True, color="blue")
plt.show()
```

#### Inference:
- The Customers Usually have the conversation between 30-40 mins.
- The Calling is Uniform between 30-40 Mins call

## Matrix Plot : Correlation Matrix of features 

```python
plt.figure(figsize=(6,6))
num_df = df.select_dtypes(include=["number"])
sns.heatmap(num_df.corr(), annot=True, cmap="coolwarm", linewidth=0.5)
plt.show()
```

#### Inference : 
- Call Duration and Satisfaction Score are Negatively correlated by 75%

## Box Plot: Satisfaction Score by Network Providers

```python
plt.figure(figsize=(8,8))
sns.boxplot(data=df, x='Network_Provider', y='Satisfaction_Score', hue='Network_Provider')
plt.show()
```

```python
#### Inference :
- Provider A : Having Less Satisfaction Score Compared to whole Satisfactory Values of Provider A
```

## Violin Plot

```python
sns.violinplot(data=df,  x='Network_Provider', y='Satisfaction_Score', hue='Network_Provider')
plt.show()
```

#### Inference :
- Narrow Shape Indicates consistent call duration.
- Long Tail suggest some customers have Significantly called.

# Used Car Price Analysis using matplotlib and seaborn
##### Dataset : Used Car listing from Car Dekho
##### Objective : Solve Business Problems Visually

#### Step 1 : Import & Configuration

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12,8)
```

#### Step 2: Load the Dataset

```python
df = pd.read_csv("./Datasets/CAR_DETAILS_FROM_CAR_DEKHO.csv")
df.head()
```

#### Step 3 : Fuel Type Distribution

```python
sns.countplot(data=df, x='fuel', hue='fuel', palette = 'Set2')
plt.title("Fuel Type distribution")
plt.legend()
plt.show()
```

##### Observations
- Petrol & Diesel are dominating the Resale Market where almost approx 2000 Cars for Petrol & Diesel each
- CNG & LPG are having very less market Share in Resale Market Not Even crossing 500 Units
- There is no Resale for Electric Cars in Indian Market.

#### Step 4 : Selling Price Distribution

```python
sns.histplot(df['selling_price'], kde=True, bins=30, color="skyblue")
plt.title("Selling Price Distribution")
plt.xlabel("Price (In Lacs INR)")
plt.show()
```

###### Observations
- The Cheaper the Car the More is the Resale .
- Most Cars are under 5 Lac Rs .
- Bulk Demand is under 2 Lacs Rupees only
- There are some good cars which have demands Increased between the price range of 2.2 Lacs & 4 Lacs.

#### Step 5 : Price By Fuel Type 

```python
sns.boxplot(data=df, x="fuel", y="selling_price", hue="fuel", palette="coolwarm")
plt.title("Price By Fuel Type")
plt.xlabel("Fuel Type")
plt.ylabel("Selling Price")

plt.show()
```

###### Observations:
- Diesel Vehicles are having Higher Selling Price as Average
- Petrol Vehicles are having a wider spread between the Vehicles.
- Diesel Vehicles have a better Resale amount in comparison to the other Fuel Type Vehicles

#### Step 6 : Price By Transmission 

```python
sns.boxplot(data=df, x="transmission", y="selling_price", hue="transmission", palette="coolwarm")
plt.title("Price By Transmission Type")
plt.xlabel("Transmission Type")
plt.ylabel("Selling Price")
plt.show()
```

###### Observations:
- Automatic Cars are Having more widespread pricing Than Manual Transmission.
- Automatic Cars are more expensive in lower ranges than the manual cars.

#### Step 7 : Price By Ownership History

```python
sns.violinplot(data=df, x="owner", y="selling_price", hue="owner", palette="Set3")
plt.title("Selling Price By Owner")
plt.show()
```

###### Observations: 
- The First Owner is having the highest Resale Value.
- Ownership history impacts Resale Value
- After Third Owner the Selling Price is between 0 to 2 Lacs only

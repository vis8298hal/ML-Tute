```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import numpy as np
import matplotlib.pyplot as plt
```

```python
x =np.array( [[1], [2], [3], [4], [5], [6]])
y = np.array([1.2, 1.9, 3,1, 10.5, 6.8])
```

```python
plt.plot([x[i][0] for i in range(len(x))], y)
deg=4
```

```python
#Get Polynomial Features with Degree
poly = PolynomialFeatures(degree=deg)
x_poly = poly.fit_transform(x)
x_poly
```

```python
model = LinearRegression()
model.fit(x_poly, y)
pred = model.predict(x_poly)
print(r2_score(y, pred))
```

```python
plt.plot([x[i][0] for i in range(len(x))], pred, color="red", label="predicted")
plt.plot([x[i][0] for i in range(len(x))], y, color="blue", label="actual")
plt.legend()
plt.show()
deg += 1
```

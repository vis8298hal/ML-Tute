# Numerical Features and Numerical Target

```python
import pandas as pd
from sklearn.feature_selection  import SelectKBest, f_regression, mutual_info_regression
```

```python
df = pd.read_csv("../../Datasets/sample_dataset.csv").iloc[:, 0:12].dropna()
```

```python
X = df.iloc[:, 0:10]
y = df.iloc[:, 10]
```

### Correlation using F Statistics

```python
selector = SelectKBest(f_regression, k=5)
selector
```

```python
selector.fit(X, y)
```

```python
selector.transform(X)
```

```python
X.columns[selector.get_support()]
```

### Correlation using Mutual Information

```python
selector = SelectKBest(mutual_info_regression, k = 5)
selector.fit(X, y)
```

```python
selector.transform(X), X.columns[selector.get_support()]
```

# Categorical Features Numerical Target

```python
from sklearn.feature_selection import SelectKBest, mutual_info_regression
from sklearn.preprocessing import OrdinalEncoder
import numpy as np
```

```python
df = pd.read_csv("../../Datasets/sample_dataset.csv")
```

```python
X= df.iloc[:, 0:10]
y = df["radius error"]
```

```python
X = OrdinalEncoder().fit_transform(X)
X
```

```python
selector = SelectKBest(lambda X,y:mutual_info_regression(X, y, discrete_features=True), k=5)
```

```python
X = np.nan_to_num(X)
y = np.nan_to_num(y)
```

```python
selector.fit_transform(X, y)
```

```python
selector = SelectKBest(lambda X,y:mutual_info_regression(X, y, discrete_features=[0, 1, 2]), k=5)
selector.fit_transform(X, y)
```

# Categorical Feature with Categorical Target

```python
from sklearn.feature_selection import chi2, mutual_info_classif
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
```

```python
df = pd.read_csv("../../Datasets/sample_dataset.csv")
X = df.iloc[:, 0:10]
y = df.iloc[:, -1]
```

```python
X = OrdinalEncoder().fit_transform(X)
y = LabelEncoder().fit_transform(y)
```

```python
X = np.nan_to_num(X)
y = np.nan_to_num(y)
```

### Mutual Information

```python
selector = SelectKBest(lambda X, y: mutual_info_classif(X, y, discrete_features=True), k=5)
selector.fit(X, y)
```

```python
selector.transform(X), [selector.get_support()]
```

### Chi Square

```python
selector = SelectKBest(chi2, k=5)
selector.fit(X, y)
```

```python
selector.transform(X)
```

# Feature Selection According to Model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.feature_selection import SelectFromModel
```

```python
df = pd.read_csv("../../Datasets/sample_dataset.csv")
df = df.loc[:, ['mean radius', 'mean texture', 'mean perimeter', 'mean area',
       'mean smoothness', 'mean compactness', 'mean concavity',
       'mean concave points', 'mean symmetry', 'mean fractal dimension', 'target']].dropna()
```

```python
X = df.iloc[:, 0:10]
y = df.iloc[:, -1]
```

```python
model = RandomForestClassifier(random_state=0)
model
```

```python
selector = SelectFromModel(model)
selector
```

```python
selector.fit_transform(X, y)
```

```python
X.columns[selector.get_support()]
```

```python
selector
```

```python
selector.estimator_.feature_importances_
```

### Limiting Number of Features

```python
selector = SelectFromModel(model, threshold="median", max_features=3)
selector
```

```python
selector.fit_transform(X, y)
```

### Selecting N most important Features

```python
selector = SelectFromModel(model, threshold=-np.inf, max_features=3)
selector
```

```python
selector.fit_transform(X, y)
```

```python
X.columns[selector.get_support()]
```

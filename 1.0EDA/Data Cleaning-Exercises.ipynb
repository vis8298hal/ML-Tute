# Exercise 1
- Load *sample_dataset.csv*
- Replace the missing categorical variables with "N"
- Replace the missing Numerical Variables with the mean Value

```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer, make_column_selector
df = pd.read_csv(r"C:\Users\kumar\Downloads\ML-Projects\Datasets\sample_dataset.csv")
cleaner = ColumnTransformer([('Numerical Features', SimpleImputer(strategy="mean"), make_column_selector(dtype_exclude="object")),
                             ('Categorical Features', SimpleImputer(strategy="constant", fill_value="N"), make_column_selector(dtype_include="object"))])
summ1 = pd.DataFrame(cleaner.fit_transform(df))
```

# Exercise 2
- Load sample_dataset.csv as dataset
- Replace missing values in the float variable using KNN with 10 neigbors and distance based weights
- Replace missing in the categorical variables using the most frequent value

```python
import pandas as pd
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.compose import ColumnTransformer, make_column_selector
df = pd.read_csv(r"C:\Users\kumar\Downloads\ML-Projects\Datasets\sample_dataset.csv")
cleaner = ColumnTransformer([('Float Features', KNNImputer(n_neighbors=10, weights="distance"), make_column_selector(dtype_include="float64")),
                             ('Categorical Features', SimpleImputer(strategy="constant", fill_value="N"), make_column_selector(dtype_include="object"))])
summ2 = pd.DataFrame(cleaner.fit_transform(df))
```

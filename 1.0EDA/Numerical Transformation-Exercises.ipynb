# Exercise 1

* Load **sample_dataset.csv**
* Consider only the first 3 columns and remove all the records that have missings
* Calculate the mean value of "mean radius"
* Use a ColumnTransformer to apply the following transformations:
    * Binarize "mean radius" using a threshold equal to the mean value
    * Binning of "mean texture" with 10 uniform bins and one-hot encoded dense output
    * Binning of "mean perimeter" with 5 quantile bins and ordinal encoding output

```python
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer, make_column_selector
from sklearn.preprocessing import Binarizer, KBinsDiscretizer
```

```python
df = pd.read_csv(r"C:\Users\kumar\Downloads\ML-Projects\Datasets\sample_dataset.csv").iloc[:,:3].dropna()
df
```

```python
m = df["mean radius"].mean()
```

```python
transformer = ColumnTransformer([('mean_radius_transform', Binarizer(threshold=m), ["mean radius"]),
                                ('mean_texture_transform', KBinsDiscretizer(strategy="uniform", n_bins=10, encode="onehot-dense"), ["mean texture"]),
                                ('mean_parameter_transform', KBinsDiscretizer(strategy="quantile", n_bins=5, encode="ordinal"), ["mean perimeter"])])
transformer.fit_transform(df)
```

# Exercise 2

* Load **sample_dataset.csv**
* Consider only the first 3 columns and remove all the records that have missings
* Use a ColumnTransformer to apply the following transformations:
    * Yeo-Johnson power transform of "mean radius"
    * Box-Cox transformation of "mean texture"
    * Calculate the logarithm of "mean perimeter"

```python
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer, make_column_selector
from sklearn.preprocessing import PowerTransformer, FunctionTransformer
```

```python
df = pd.read_csv(r"C:\Users\kumar\Downloads\ML-Projects\Datasets\sample_dataset.csv").iloc[:,:3].dropna()
df
```

```python
transformer = ColumnTransformer([('mean_radius_transform', PowerTransformer(), ["mean radius"]),
                                ('mean_texture_transform', PowerTransformer('box-cox'), ["mean texture"]),
                                ('mean_parameter_transform', FunctionTransformer(np.log, validate=True), ["mean perimeter"])])
transformer.fit_transform(df)
```

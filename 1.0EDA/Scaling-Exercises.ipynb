- Numerical Variables must be cleaned using the median value then normalized
- Change the Scaler to Standard Scaler and transform dataset again

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer, make_column_selector
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv("../../Datasets/sample_dataset.csv")
```

```python
num_pipe = Pipeline([
    ('cleaner', SimpleImputer(strategy="median")),
    ('scaler', MinMaxScaler())
])
```

```python
transformer = ColumnTransformer([
    ('numerical', num_pipe, make_column_selector(dtype_exclude="object"))
])
transformer
```

```python
transformer.fit_transform(df)
```

```python
transformer.set_params(numerical__scaler = StandardScaler())
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("../../Datasets/sample_dataset.csv")
```

```python
X = df.iloc[:,0:10].dropna()
```

```python
pca = PCA()
```

```python
pca.fit_transform(X)
```

# Dimensionality Reduction

```python
pca = PCA(n_components=5)
```

```python
print(pca.fit_transform(X))
```

# Dimensionality Reduction with Variance filter 

```python
pca.explained_variance_
```

```python
pca = PCA(n_components=9.86721005e-01)
pca.fit_transform(X)
```

# Selecting best number of components

```python
import matplotlib.pyplot as plt
```

```python
pca = PCA()
pca.fit(X)
```

```python
x = range(1, len(pca.explained_variance_ratio_)+1)
y = pca.explained_variance_ratio_
```

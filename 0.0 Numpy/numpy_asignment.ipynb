# 1: Create a 1D NumPy array with integers from 1 to 20. Perform the following operations:

```python
arr = np.arange(1,21)
```

### Calculate the sum, mean, median, and standard deviation of the elements in the array.  


```python
print(f"Sum : {np.sum(arr)}")
print(f"Mean : {np.mean(arr)}")
print(f"Median : {np.median(arr)}")
print(f"Standard Deviation : {np.std(arr)}")
```

### Find the indices of elements greater than 10 in the array.

```python
print("The Indices for Elements Greater than 10 are")
for i in np.where(arr > 10)[0]:
    print(i, end=" ")
```

# 2. Create a 2D NumPy array of shape 4 X 4 with numbers ranging from 1 to 16.

```python
arr2d = np.arange(1,17).reshape(4,4)
```

### Print the array.  

```python
print(arr2d)
```

 ### Find the transpose of the array. 

```python
arr2dTra = np.transpose(arr2d)
print(arr2dTra)
```

### Calculate the row-wise and column-wise sums of the array.

```python
print("Row Wise Sum : ",np.sum(arr2d, axis=1))
print("Column Wise Sum : ", np.sum(arr2d, axis=0))
```

# 3. Create two 3 X 3 arrays filled with random integers between 1 and 20.

```python
arr1 = np.random.randint(1, 20, size=9).reshape(3,3)
arr2 = np.random.randint(1, 20, size=9).reshape(3,3)
```

### Perform element-wise addition, subtraction, and multiplication

```python
print("Array 1 : ")
print(arr1)
print("Array 2 : ")
print(arr2)
```

```python
print("Element wise sum of the Arrays : ")
print(arr1 + arr2)
```

```python
print("Element wise Subtraction")
print(arr1 - arr2)
```

```python
print("Element wise multiplication")
print(arr1 * arr2)
```

### Compute the dot product of the two arrays

```python
print(np.dot(arr1, arr2))
```
